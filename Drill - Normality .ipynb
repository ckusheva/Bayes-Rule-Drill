```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
var = np.random.normal(2, 10, 1000)
mean = np.mean(var)
sd = np.std(var)
plt.hist(var)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show
```
